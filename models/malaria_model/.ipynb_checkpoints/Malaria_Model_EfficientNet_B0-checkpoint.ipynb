```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import copy

# Set random seeds
torch.manual_seed(42)
np.random.seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)

print("="*80)
print("MALARIA CELL CLASSIFICATION - EFFICIENTNET-B0 TRAINING (PYTORCH)")
print("="*80)
print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA Available: {torch.cuda.is_available()}")

# ============================================================================
# 1. CONFIGURATION
# ============================================================================

# Paths
train_dir = r"C:\Users\USER\honors-final-project\models\malaria_dataset\split_data\train"
val_dir = r"C:\Users\USER\honors-final-project\models\malaria_dataset\split_data\val"
model_save_path = r"C:\Users\USER\honors-final-project\models\malaria_dataset\best_malaria_model.pt"

# Hyperparameters
IMG_SIZE = 224
BATCH_SIZE = 128
EPOCHS = 50
LEARNING_RATE = 0.001
EARLY_STOP_PATIENCE = 10

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

print(f"\nConfiguration:")
print(f"  Device: {device}")
print(f"  Image Size: {IMG_SIZE}x{IMG_SIZE}")
print(f"  Batch Size: {BATCH_SIZE}")
print(f"  Max Epochs: {EPOCHS}")
print(f"  Learning Rate: {LEARNING_RATE}")
print(f"  Early Stopping Patience: {EARLY_STOP_PATIENCE}")


# ============================================================================
# 2. DATA PREPARATION
# ============================================================================

print("\n" + "="*80)
print("DATA PREPARATION")
print("="*80)

# Data augmentation for training
train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomRotation(20),
    transforms.RandomHorizontalFlip(),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Only normalization for validation
val_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
train_dataset = datasets.ImageFolder(train_dir, transform=train_transform)
val_dataset = datasets.ImageFolder(val_dir, transform=val_transform)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4)

print(f"\nTraining samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")
print(f"Class names: {train_dataset.classes}")
print(f"Class indices: {train_dataset.class_to_idx}")


# ============================================================================
# 3. BUILD MODEL
# ============================================================================

print("\n" + "="*80)
print("MODEL ARCHITECTURE")
print("="*80)

# Load EfficientNet-B0 with pretrained weights
model = models.efficientnet_b0(pretrained=True)

# Freeze base layers
for param in model.parameters():
    param.requires_grad = False

# Modify classifier for binary classification
num_features = model.classifier[1].in_features
model.classifier = nn.Sequential(
    nn.Dropout(0.3),
    nn.Linear(num_features, 128),
    nn.ReLU(),
    nn.BatchNorm1d(128),
    nn.Dropout(0.3),
    nn.Linear(128, 1),
    nn.Sigmoid()
)

model = model.to(device)

print(f"\nModel loaded: EfficientNet-B0")
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")


# ============================================================================
# 4. LOSS & OPTIMIZER
# ============================================================================

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)


# ============================================================================
# 5. TRAINING FUNCTION
# ============================================================================

def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for inputs, labels in loader:
        inputs, labels = inputs.to(device), labels.to(device).float().unsqueeze(1)
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * inputs.size(0)
        predicted = (outputs > 0.5).float()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


def validate_epoch(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device).float().unsqueeze(1)
            
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item() * inputs.size(0)
            predicted = (outputs > 0.5).float()
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


# ============================================================================
# 6. TRAINING LOOP
# ============================================================================

print("\n" + "="*80)
print("STARTING TRAINING")
print("="*80)

# Initialize tracking variables
history = {
    'train_loss': [], 'train_acc': [],
    'val_loss': [], 'val_acc': []
}

best_val_acc = 0.0
best_model_wts = copy.deepcopy(model.state_dict())
epochs_no_improve = 0

# Start timer
start_time = time.time()

# Training loop
for epoch in range(EPOCHS):
    # Train
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    
    # Validate
    val_loss, val_acc = validate_epoch(model, val_loader, criterion, device)
    
    # Store history
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)
    
    # Print epoch details
    print(f"\nEpoch {epoch + 1}/{EPOCHS}")
    print(f"  Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"  Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")
    
    # Check if this is the best model
    if val_acc > best_val_acc:
        print(f"  ✓ Validation accuracy improved from {best_val_acc:.4f} to {val_acc:.4f}")
        print(f"  ✓ Saving best model to {model_save_path}")
        best_val_acc = val_acc
        best_model_wts = copy.deepcopy(model.state_dict())
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_acc': val_acc,
            'val_loss': val_loss,
        }, model_save_path)
        epochs_no_improve = 0
    else:
        epochs_no_improve += 1
        print(f"  No improvement for {epochs_no_improve} epoch(s)")
    
    print("-" * 80)
    
    # Early stopping
    if epochs_no_improve >= EARLY_STOP_PATIENCE:
        print(f"\n⚠ Early stopping triggered after {epoch + 1} epochs")
        print(f"No improvement in validation accuracy for {EARLY_STOP_PATIENCE} consecutive epochs")
        break

# End timer
end_time = time.time()
training_time = end_time - start_time

# Load best model
model.load_state_dict(best_model_wts)

print("\n" + "="*80)
print("TRAINING COMPLETED")
print("="*80)
print(f"Total Training Time: {training_time/60:.2f} minutes ({training_time:.2f} seconds)")
print(f"Best Validation Accuracy: {best_val_acc:.4f}")
print(f"Best Model saved at: {model_save_path}")


# ============================================================================
# 7. EVALUATION - GET PREDICTIONS
# ============================================================================

print("\n" + "="*80)
print("EVALUATION ON VALIDATION SET")
print("="*80)

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.to(device)
        outputs = model(inputs)
        predicted = (outputs > 0.5).float().cpu().numpy()
        all_preds.extend(predicted)
        all_labels.extend(labels.numpy())

y_true = np.array(all_labels)
y_pred = np.array(all_preds).flatten().astype(int)


# ============================================================================
# 8. CONFUSION MATRIX
# ============================================================================

print("\n" + "="*80)
print("CONFUSION MATRIX")
print("="*80)

cm = confusion_matrix(y_true, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=train_dataset.classes, 
            yticklabels=train_dataset.classes,
            cbar_kws={'label': 'Count'})
plt.title('Confusion Matrix', fontsize=16, fontweight='bold')
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.show()


# ============================================================================
# 9. CLASSIFICATION REPORT
# ============================================================================

print("\n" + "="*80)
print("CLASSIFICATION REPORT")
print("="*80)

report = classification_report(y_true, y_pred, 
                               target_names=train_dataset.classes, 
                               digits=4)
print("\n" + report)


# ============================================================================
# 10. TRAINING HISTORY PLOTS
# ============================================================================

print("\n" + "="*80)
print("TRAINING HISTORY VISUALIZATION")
print("="*80)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1: Accuracy
axes[0].plot(history['train_acc'], label='Train Accuracy', marker='o', linewidth=2)
axes[0].plot(history['val_acc'], label='Val Accuracy', marker='s', linewidth=2)
axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Accuracy', fontsize=12)
axes[0].legend(loc='lower right', fontsize=10)
axes[0].grid(alpha=0.3)

# Plot 2: Loss
axes[1].plot(history['train_loss'], label='Train Loss', marker='o', linewidth=2)
axes[1].plot(history['val_loss'], label='Val Loss', marker='s', linewidth=2)
axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Loss', fontsize=12)
axes[1].legend(loc='upper right', fontsize=10)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()


# ============================================================================
# 11. FINAL SUMMARY
# ============================================================================

print("\n" + "="*80)
print("FINAL SUMMARY")
print("="*80)

final_train_acc = history['train_acc'][-1]
final_val_acc = history['val_acc'][-1]
final_train_loss = history['train_loss'][-1]
final_val_loss = history['val_loss'][-1]

best_val_acc_epoch = history['val_acc'].index(max(history['val_acc'])) + 1

print(f"\nTraining Time: {training_time/60:.2f} minutes")
print(f"Total Epochs Run: {len(history['train_acc'])}")
print(f"Best Validation Accuracy: {best_val_acc:.4f} (Epoch {best_val_acc_epoch})")
print(f"\nFinal Metrics:")
print(f"  Train Accuracy: {final_train_acc:.4f}")
print(f"  Train Loss: {final_train_loss:.4f}")
print(f"  Val Accuracy: {final_val_acc:.4f}")
print(f"  Val Loss: {final_val_loss:.4f}")

print("\n" + "="*80)
print("ALL DONE! ✓")
print("="*80)
```

### Fine Tuning

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import copy

# Set random seeds
torch.manual_seed(42)
np.random.seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)

print("="*80)
print("MALARIA CELL CLASSIFICATION - FINE-TUNING")
print("="*80)
print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA Available: {torch.cuda.is_available()}")

# ============================================================================
# 1. CONFIGURATION
# ============================================================================

# Paths
train_dir = r"C:\Users\USER\honors-final-project\models\malaria_dataset\split_data\train"
val_dir = r"C:\Users\USER\honors-final-project\models\malaria_dataset\split_data\val"
pretrained_model_path = r"C:\Users\USER\honors-final-project\models\malaria_dataset\best_malaria_model.pt"
finetuned_model_path = r"C:\Users\USER\honors-final-project\models\malaria_dataset\best_malaria_model_finetuned.pt"

# Hyperparameters for fine-tuning
IMG_SIZE = 224
BATCH_SIZE = 64  # Reduced for fine-tuning stability
EPOCHS = 30
LEARNING_RATE = 0.0001  # Lower learning rate for fine-tuning
EARLY_STOP_PATIENCE = 8

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

print(f"\nFine-tuning Configuration:")
print(f"  Device: {device}")
print(f"  Image Size: {IMG_SIZE}x{IMG_SIZE}")
print(f"  Batch Size: {BATCH_SIZE}")
print(f"  Max Epochs: {EPOCHS}")
print(f"  Learning Rate: {LEARNING_RATE} (lower for fine-tuning)")
print(f"  Early Stopping Patience: {EARLY_STOP_PATIENCE}")


# ============================================================================
# 2. DATA PREPARATION
# ============================================================================

print("\n" + "="*80)
print("DATA PREPARATION")
print("="*80)

# Training data with stronger augmentation
train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomRotation(25),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(p=0.3),
    transforms.RandomAffine(degrees=0, translate=(0.15, 0.15)),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.2),
    transforms.RandomPerspective(distortion_scale=0.2, p=0.5),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Validation data
val_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
train_dataset = datasets.ImageFolder(train_dir, transform=train_transform)
val_dataset = datasets.ImageFolder(val_dir, transform=val_transform)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4)

print(f"\nTraining samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")
print(f"Class names: {train_dataset.classes}")


# ============================================================================
# 3. LOAD PRETRAINED MODEL AND UNFREEZE LAYERS
# ============================================================================

print("\n" + "="*80)
print("MODEL LOADING & UNFREEZING")
print("="*80)

# Load base EfficientNet-B0
model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)

# Recreate the classifier (same as before)
num_features = model.classifier[1].in_features
model.classifier = nn.Sequential(
    nn.Dropout(0.3),
    nn.Linear(num_features, 128),
    nn.ReLU(),
    nn.BatchNorm1d(128),
    nn.Dropout(0.3),
    nn.Linear(128, 1),
    nn.Sigmoid()
)

model = model.to(device)

# Load the pretrained weights
print(f"\nLoading pretrained model from: {pretrained_model_path}")
checkpoint = torch.load(pretrained_model_path)
model.load_state_dict(checkpoint['model_state_dict'])
print(f"✓ Loaded model with validation accuracy: {checkpoint['val_acc']:.4f}")

# Unfreeze last 3 feature blocks for fine-tuning
print("\nUnfreezing layers for fine-tuning...")
for param in model.parameters():
    param.requires_grad = False

# Unfreeze last 3 blocks of features
for param in model.features[-3:].parameters():
    param.requires_grad = True

# Keep classifier trainable
for param in model.classifier.parameters():
    param.requires_grad = True

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTotal parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
print(f"Frozen parameters: {total_params - trainable_params:,}")


# ============================================================================
# 4. OPTIMIZER & SCHEDULER
# ============================================================================

criterion = nn.BCELoss()

# Different learning rates for different parts
optimizer = optim.Adam([
    {'params': model.features[-3:].parameters(), 'lr': LEARNING_RATE * 0.1},  # Lower LR for base
    {'params': model.classifier.parameters(), 'lr': LEARNING_RATE}  # Higher LR for classifier
])

# Learning rate scheduler
from torch.optim.lr_scheduler import ReduceLROnPlateau
scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=3, verbose=True)


# ============================================================================
# 5. TRAINING FUNCTIONS
# ============================================================================

def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for inputs, labels in loader:
        inputs, labels = inputs.to(device), labels.to(device).float().unsqueeze(1)
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * inputs.size(0)
        predicted = (outputs > 0.5).float()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


def validate_epoch(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device).float().unsqueeze(1)
            
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item() * inputs.size(0)
            predicted = (outputs > 0.5).float()
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


# ============================================================================
# 6. FINE-TUNING LOOP
# ============================================================================

print("\n" + "="*80)
print("STARTING FINE-TUNING")
print("="*80)

# Initialize tracking variables
history = {
    'train_loss': [], 'train_acc': [],
    'val_loss': [], 'val_acc': []
}

best_val_acc = checkpoint['val_acc']  # Start from previous best
best_model_wts = copy.deepcopy(model.state_dict())
epochs_no_improve = 0

print(f"Starting from validation accuracy: {best_val_acc:.4f}")

# Start timer
start_time = time.time()

# Training loop
for epoch in range(EPOCHS):
    # Train
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    
    # Validate
    val_loss, val_acc = validate_epoch(model, val_loader, criterion, device)
    
    # Store history
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)
    
    # Print epoch details
    print(f"\nEpoch {epoch + 1}/{EPOCHS}")
    print(f"  Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"  Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")
    
    # Learning rate scheduler step
    scheduler.step(val_acc)
    current_lr = optimizer.param_groups[0]['lr']
    print(f"  Learning Rate: {current_lr:.6f}")
    
    # Check if this is the best model
    if val_acc > best_val_acc:
        improvement = val_acc - best_val_acc
        print(f"  ✓ Validation accuracy improved from {best_val_acc:.4f} to {val_acc:.4f} (+{improvement:.4f})")
        print(f"  ✓ Saving best model to {finetuned_model_path}")
        best_val_acc = val_acc
        best_model_wts = copy.deepcopy(model.state_dict())
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_acc': val_acc,
            'val_loss': val_loss,
        }, finetuned_model_path)
        epochs_no_improve = 0
    else:
        epochs_no_improve += 1
        print(f"  No improvement for {epochs_no_improve} epoch(s)")
    
    print("-" * 80)
    
    # Early stopping
    if epochs_no_improve >= EARLY_STOP_PATIENCE:
        print(f"\n⚠ Early stopping triggered after {epoch + 1} epochs")
        print(f"No improvement in validation accuracy for {EARLY_STOP_PATIENCE} consecutive epochs")
        break

# End timer
end_time = time.time()
training_time = end_time - start_time

# Load best model
model.load_state_dict(best_model_wts)

print("\n" + "="*80)
print("FINE-TUNING COMPLETED")
print("="*80)
print(f"Total Fine-tuning Time: {training_time/60:.2f} minutes ({training_time:.2f} seconds)")
print(f"Best Validation Accuracy: {best_val_acc:.4f}")
print(f"Improvement from initial: {best_val_acc - checkpoint['val_acc']:.4f}")
print(f"Fine-tuned Model saved at: {finetuned_model_path}")


# ============================================================================
# 7. EVALUATION - GET PREDICTIONS
# ============================================================================

print("\n" + "="*80)
print("EVALUATION ON VALIDATION SET")
print("="*80)

model.eval()
all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.to(device)
        outputs = model(inputs)
        predicted = (outputs > 0.5).float().cpu().numpy()
        all_preds.extend(predicted)
        all_labels.extend(labels.numpy())
        all_probs.extend(outputs.cpu().numpy())

y_true = np.array(all_labels)
y_pred = np.array(all_preds).flatten().astype(int)
y_probs = np.array(all_probs).flatten()


# ============================================================================
# 8. CONFUSION MATRIX
# ============================================================================

print("\n" + "="*80)
print("CONFUSION MATRIX")
print("="*80)

cm = confusion_matrix(y_true, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Calculate per-class metrics
tn, fp, fn, tp = cm.ravel()
print(f"\nTrue Negatives (TN): {tn}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")
print(f"True Positives (TP): {tp}")

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=train_dataset.classes, 
            yticklabels=train_dataset.classes,
            cbar_kws={'label': 'Count'},
            annot_kws={'size': 16})
plt.title('Confusion Matrix - Fine-tuned Model', fontsize=18, fontweight='bold', pad=20)
plt.ylabel('True Label', fontsize=14)
plt.xlabel('Predicted Label', fontsize=14)
plt.tight_layout()
plt.show()


# ============================================================================
# 9. CLASSIFICATION REPORT
# ============================================================================

print("\n" + "="*80)
print("CLASSIFICATION REPORT")
print("="*80)

report = classification_report(y_true, y_pred, 
                               target_names=train_dataset.classes, 
                               digits=4)
print("\n" + report)

# Calculate additional metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
auc_score = roc_auc_score(y_true, y_probs)

print("\n" + "="*80)
print("ADDITIONAL METRICS")
print("="*80)
print(f"\nOverall Metrics:")
print(f"  Accuracy:  {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"  Precision: {precision:.4f}")
print(f"  Recall:    {recall:.4f}")
print(f"  F1-Score:  {f1:.4f}")
print(f"  AUC-ROC:   {auc_score:.4f}")

# Specificity and Sensitivity
sensitivity = tp / (tp + fn)  # Same as recall
specificity = tn / (tn + fp)
print(f"\nClinical Metrics:")
print(f"  Sensitivity (True Positive Rate):  {sensitivity:.4f}")
print(f"  Specificity (True Negative Rate):  {specificity:.4f}")


# ============================================================================
# 10. ROC CURVE
# ============================================================================

from sklearn.metrics import roc_curve

print("\n" + "="*80)
print("ROC CURVE")
print("="*80)

fpr, tpr, thresholds = roc_curve(y_true, y_probs)

plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc_score:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=16, fontweight='bold')
plt.legend(loc="lower right", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


# ============================================================================
# 11. TRAINING HISTORY PLOTS
# ============================================================================

print("\n" + "="*80)
print("TRAINING HISTORY VISUALIZATION")
print("="*80)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Accuracy
axes[0].plot(history['train_acc'], label='Train Accuracy', marker='o', linewidth=2, markersize=6)
axes[0].plot(history['val_acc'], label='Val Accuracy', marker='s', linewidth=2, markersize=6)
axes[0].axhline(y=checkpoint['val_acc'], color='r', linestyle='--', 
                label=f'Pre-finetuning ({checkpoint["val_acc"]:.4f})', linewidth=2)
axes[0].set_title('Model Accuracy - Fine-tuning', fontsize=16, fontweight='bold')
axes[0].set_xlabel('Epoch', fontsize=13)
axes[0].set_ylabel('Accuracy', fontsize=13)
axes[0].legend(loc='lower right', fontsize=11)
axes[0].grid(alpha=0.3)

# Plot 2: Loss
axes[1].plot(history['train_loss'], label='Train Loss', marker='o', linewidth=2, markersize=6)
axes[1].plot(history['val_loss'], label='Val Loss', marker='s', linewidth=2, markersize=6)
axes[1].set_title('Model Loss - Fine-tuning', fontsize=16, fontweight='bold')
axes[1].set_xlabel('Epoch', fontsize=13)
axes[1].set_ylabel('Loss', fontsize=13)
axes[1].legend(loc='upper right', fontsize=11)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()


# ============================================================================
# 12. COMPARISON TABLE
# ============================================================================

print("\n" + "="*80)
print("BEFORE vs AFTER FINE-TUNING COMPARISON")
print("="*80)

comparison_data = {
    'Metric': ['Validation Accuracy', 'Validation Loss', 'Training Time', 'Trainable Params'],
    'Before Fine-tuning': [
        f"{checkpoint['val_acc']:.4f}",
        f"{checkpoint['val_loss']:.4f}",
        "21.86 min",
        "164,353"
    ],
    'After Fine-tuning': [
        f"{best_val_acc:.4f}",
        f"{history['val_loss'][history['val_acc'].index(max(history['val_acc']))]:.4f}",
        f"{training_time/60:.2f} min",
        f"{trainable_params:,}"
    ],
    'Improvement': [
        f"{(best_val_acc - checkpoint['val_acc']):.4f}",
        f"{(checkpoint['val_loss'] - history['val_loss'][history['val_acc'].index(max(history['val_acc']))]):.4f}",
        "-",
        "-"
    ]
}

import pandas as pd
comparison_df = pd.DataFrame(comparison_data)
print("\n" + comparison_df.to_string(index=False))


# ============================================================================
# 13. FINAL SUMMARY
# ============================================================================

print("\n" + "="*80)
print("FINAL SUMMARY")
print("="*80)

final_train_acc = history['train_acc'][-1]
final_val_acc = history['val_acc'][-1]
final_train_loss = history['train_loss'][-1]
final_val_loss = history['val_loss'][-1]

best_val_acc_epoch = history['val_acc'].index(max(history['val_acc'])) + 1

print(f"\nFine-tuning Time: {training_time/60:.2f} minutes")
print(f"Total Epochs Run: {len(history['train_acc'])}")
print(f"Best Validation Accuracy: {best_val_acc:.4f} (Epoch {best_val_acc_epoch})")
print(f"Improvement: +{(best_val_acc - checkpoint['val_acc'])*100:.2f}%")

print(f"\nFinal Metrics:")
print(f"  Train Accuracy: {final_train_acc:.4f}")
print(f"  Train Loss: {final_train_loss:.4f}")
print(f"  Val Accuracy: {final_val_acc:.4f}")
print(f"  Val Loss: {final_val_loss:.4f}")

print(f"\nModel Performance:")
print(f"  Correctly Classified: {tp + tn} / {len(y_true)} ({accuracy*100:.2f}%)")
print(f"  Misclassified: {fp + fn} / {len(y_true)} ({(1-accuracy)*100:.2f}%)")

print("\n" + "="*80)
print("FINE-TUNING COMPLETE! ✓")
print("="*80)
```
